## Q2: How to identify which country has the most VIP Customers?

```python
country_amount = data.groupby(['CustomerID','Country']).agg({'amount_spent':sum})
perc95 = np.percentile(agg_amount['amount_spent'],95)
perc75 = np.percentile(agg_amount['amount_spent'],75)

def customers_def(x):
    if x > perc95:
        return 'VIP Customer'
    elif x < perc95 and x >= perc75:
        return 'Preferred Customer'
    else:
        return 'Not-preferred Customer'
    
country_amount['customer_label'] = country_amount['amount_spent'].apply(customers_def)

country_amount
```

```python

country_amount['customer_label'].value_counts()
```

```python
country_amount = country_amount.reset_index()
country_amount
```

```python
only_vip = country_amount[country_amount['customer_label'].isin(['VIP Customer','Preferred Customer'])]
only_vip
```

```python
only_vip['Country'].value_counts()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
vip_and_pref = country_amount[country_amount['customer_label'].isin(['VIP Customer','Preferred Customer'])]
vip_and_pref
```
